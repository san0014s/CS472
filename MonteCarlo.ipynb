```python
%matplotlib inline
from scipy import random
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = np.pi #limits of integration
N = 1000
randarr = np.zeros(N)

randarr = random.uniform(a,b,N)

def func(x):
    return np.sin(x)

integral = 0.0 #starting out

#for i in range(N):
    #integral += func(randarr[i]) #evaluating the function at the next value

answer = (b-a)/float (N)*integral

areas = []

#print(answer) #integral from 0 to pi of sin(x)
for i in range(N):
    randarr = np.zeros(N)

    for i in range(len(randarr)):
        randarr[i] = random.uniform(a,b)
        integral = 0.0

    for i in range(N):
        integral += func(randarr[i])

                       
    answer = (b-a)/float(N)*integral
    areas.append(answer)
    
plt.title("Distribution of Areas Calculated")
plt.hist(areas,bins = 30, ec = 'black')
plt.xlabel("Areas")
```
